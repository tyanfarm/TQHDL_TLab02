```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('.\Data\processed_data_kansas.csv')

df
```

# Sự biến đổi của tội phạm có sử dụng súng và không sử dụng súng trong năm 2016

```python

import plotly.express as px

# Đảm bảo cột 'From_Date' có định dạng datetime
df['From_Date'] = pd.to_datetime(df['From_Date'])

# Tạo một cột mới để lưu ngày mà không có thông tin về thời gian
df['Date'] = df[df['From_Date'].dt.year == 2016]['From_Date'].dt.date

# Tạo một cột mới để chỉ ra việc sử dụng súng hay không
df['Firearm_Used'] = df['Firearm Used Flag'].apply(lambda x: 'Có sử dụng' if x == 'Y' else 'Không sử dụng')

# Lọc dữ liệu cho năm 2016
df_2016 = df[df['From_Date'].dt.year == 2016]

# Nhóm theo Ngày và Firearm_Used
grouped_data = df_2016.groupby(['Date', 'Firearm_Used']).size().reset_index(name='Total_Crimes')

# Tạo biểu đồ
fig = px.line(grouped_data, x='Date', y='Total_Crimes', color='Firearm_Used', 
              title='Số Vụ Phạm Tội Hàng Ngày ở Kansas City Có hoặc Không có Sử Dụng Súng',
              labels={'Total_Crimes': 'Tổng số vụ Phạm Tội', 'Date': 'Thời gian', 'Firearm_Used': 'Sử dụng súng'})
fig.show()

```

**Xu hướng:**

* **Số vụ phạm tội không sử dụng súng:**  Có xu hướng tăng từ tháng 1 đến tháng 7, đạt đỉnh điểm vào tháng 7, sau đó giảm dần đến tháng 11.
* **Số vụ phạm tội có sử dụng súng:**  Không có xu hướng rõ ràng, dao động bất thường trong suốt thời gian khảo sát.

**Giao động:**
* **Số vụ phạm tội không sử dụng súng:**  Dạo động lớn, có thể có giao động từ 200 lên 400 trong 1 khoảng thời gian ngắn.
* **Số vụ phạm tội có sử dụng súng:**  Dao động nhỏ hơn, chỉ có thể nhảy vọt từ 0 đến 80 trong 1 khoảng thời gian ngắn.

**Đỉnh cao thấp:**

* **Số vụ phạm tội không sử dụng súng:** Đỉnh cao nhất trong khoảng tháng 7, thấp nhất vào khoảng tháng 1.
* **Số vụ phạm tội có sử dụng súng:**  Đỉnh cao nhất vào tháng 5 và tháng 9, thấp nhất vào tháng 2 và tháng 10.

**Kết luận:**

- Số vụ **phạm tội không sử dụng súng** có thể có xu hướng theo mùa, tăng vào mùa hè và giảm vào mùa thu.
- Số vụ **phạm tội có sử dụng súng** không có xu hướng rõ ràng, dao động bất thường hơn.
- Nhìn chung, số vụ **phạm tội không sử dụng súng** nhiều hơn hẳn so với số vụ **phạm tội có sử dụng súng**.


# Số lượng tội phạm Hàng ngày cho top 3 loại tội phạm trong năm 2016

### Thực hiện lấy top 3 tội phạm.

```python
# Tạo DataFrame từ dữ liệu top loại tội phạm
  # Số lượng top loại tội phạm bạn muốn hiển thị
top_crimes_data = df['Description'].value_counts().head(3)
```

### Xử lí dữ liệu và vẽ biểu đồ

```python


# Đảm bảo cột 'From_Date' có định dạng datetime
df['From_Date'] = pd.to_datetime(df['From_Date'])

# Lọc dữ liệu cho năm 2016 và các loại tội phạm trong top
df_2016_top_crimes = df[(df['From_Date'].dt.year == 2016) & (df['Description'].isin(top_crimes_data.index))]

# Group by From_Date and Description
grouped_data = df_2016_top_crimes.groupby(['From_Date', 'Description']).size().reset_index(name='Total_Crimes')

# Tạo biểu đồ
fig = px.line(grouped_data, x='From_Date', y='Total_Crimes', color='Description', 
              title=f'Số Lượng Tội Phạm Hàng Ngày cho Top {3} Loại Tội Phạm trong Năm 2016',
              labels={'Total_Crimes': 'Số Lượng Tội Phạm', 'From_Date': 'Thời gian', 'Description': 'Loại Tội Phạm'})
fig.show()
```

**Xu hướng:**

* **Burglary - Residence:** Có xu hướng tăng nhẹ từ tháng 1 đến tháng 3, sau đó giảm dần đến tháng 5, tăng trở lại vào tháng 7, đạt đỉnh vào tháng 9 và giảm dần đến cuối năm.
* **Non Agg Assault Dome:** Có xu hướng dao động mạnh hơn, không có xu hướng rõ ràng. Tăng đột biến vào tháng 2, tháng 4 và tháng 7, sau đó giảm dần và tăng trở lại vào tháng 10. 
* **Property Damage:** Có xu hướng tăng nhẹ từ tháng 1 đến tháng 3, giảm dần đến tháng 5, tăng mạnh vào tháng 6, đạt đỉnh vào tháng 7, giảm dần đến cuối năm.

**Giao động:**

* **Burglary - Residence:** Giao động tương đối nhẹ, thường nằm trong khoảng từ 10 đến 40 vụ.
* **Non Agg Assault Dome:** Giao động lớn hơn, có thể nhảy vọt từ 10 đến 50 vụ chỉ một khoảng thời gian ngắn. 
* **Property Damage:** Giao động mạnh nhất, có thể tăng giảm đột ngột trong một thời gian ngắn, từ 10 đến 60 vụ.

**Đỉnh cao thấp:**

* **Burglary - Residence:** Đỉnh cao nhất vào tháng 9 với 57 vụ, thấp nhất vào tháng 6 với 2 vụ.
* **Non Agg Assault Dome:** Đỉnh cao nhất vào tháng 7, tháng 10, thấp nhất vào tháng 2 và tháng 8.
* **Property Damage:** Đỉnh cao nhất vào tháng 7, thấp nhất vào tháng 2.

**Kết luận:**

- Các loại tội phạm đều có xu hướng tăng vào mùa hè và giảm vào mùa thu, nhưng mức độ dao động và thời điểm đạt đỉnh điểm khác nhau.
- Tội phạm "Property Damage" có biến động mạnh nhất, theo sau là "Non Agg Assault Dome", trong khi "Burglary - Residence" có biến động nhẹ nhất.
- Tội phạm "Property Damage" có số vụ cao nhất trong suốt cả năm.

#### Kiểm tra số dòng dữ liệu trùng lắp

```python
num_duplicated_rows = sum(raw_df.duplicated())
if num_duplicated_rows == 0:
    print(f"Dữ liệu thô không có dòng trùng lắp.")
else:
    print(f"Dữ liệu thô có {num_duplicated_rows} dòng trùng lắp. Cần loại bỏ các dòng này.")
```

```python
raw_df['Report_No'] = raw_df['Report_No'].astype('object')
raw_df['Offense'] = raw_df['Offense'].astype('object')
raw_df['Beat'] = raw_df['Beat'].astype('object')
raw_df['Zip Code'] = raw_df['Zip Code'].astype('object')
raw_df['Invl_No'] = raw_df['Invl_No'].astype('object')
dtypes = raw_df.dtypes
float_cols = set(dtypes[(dtypes==np.float32) | (dtypes==np.float64)].index)
object_cols = set(dtypes[dtypes == object].index)
dtypes
```

#### Sự phân bố giá trị của các cột dữ liệu kiểu số

```python
num_col_info_df = raw_df.select_dtypes(exclude='object')
def missing_ratio(df: pd.DataFrame):
    ratio = df.isnull().sum()/float(len(df)) *100
    return ratio.round(1)

def lower_quartile(df: pd.DataFrame):
    return df.quantile(0.25).round(1)


def median(df: pd.DataFrame):
    return df.quantile(0.5).round(1)


def upper_quartile(df: pd.DataFrame):
    return df.quantile(0.75).round(1)



num_col_info_df = num_col_info_df.agg([missing_ratio, "min", lower_quartile, median, upper_quartile, "max"])
num_col_info_df
```

### Loại bỏ cột, dòng dữ liệu
Trong các cột dữ liệu có cột Age có phân bố không phù hợp với thực tế (giá trị max lên tới 404) nên ta sẽ tiến hành loại bỏ nó

```python
raw_df.drop(['Age'], axis=1, inplace=True)
raw_df.head()
```

```python
raw_df.select_dtypes(include='object')
num_col_info_df = raw_df.agg([missing_ratio])
num_col_info_df
```

Hai cột To_Date và To_Time có tỉ lệ mất mát khá lớn nên ta sẽ lần lượt loại bỏ nó.

```python
raw_df.drop(['To_Date'], axis=1, inplace=True)
raw_df.head()
```

```python
raw_df.drop(['To_Time'], axis=1, inplace=True)
raw_df.head()
```

Một vài dòng có City không cụ thể (bằng giá trị '#NAME?') nên ta sẽ tiến hành xóa

```python
idx_drop_list = raw_df[raw_df['City'] == '#NAME?'].index
raw_df = raw_df.drop(axis='index', index = idx_drop_list)
raw_df = raw_df.reset_index(drop=True)
raw_df[:20]
```

Các cột IBRS, Beat, Zip Code, Rep_Dist, Area, có tỉ lệ trống không nhiều nên ta sẽ điền giá trị None vào.

```python
raw_df.fillna({'IBRS': 'None'}, inplace=True)
raw_df.agg([missing_ratio])['IBRS']
```

```python
raw_df.fillna({'Beat': 'None'}, inplace=True)
raw_df.agg([missing_ratio])['Beat']
```

```python
raw_df.fillna({'Zip Code': 'None'}, inplace=True)
raw_df.agg([missing_ratio])['Zip Code']
```

```python
raw_df.fillna({'Rep_Dist': 'None'}, inplace=True)
raw_df.agg([missing_ratio])['Rep_Dist']
```

```python
raw_df.fillna({'Area': 'None'}, inplace=True)
raw_df.agg([missing_ratio])['Area']
```

Tương tự, ta sẽ điền các giá trị trống cột Race và Sex là U, đại diện cho Unknown do chúng đã tồn tại giá trị này trong dữ liệu.

```python
raw_df.fillna({'Race': 'U'}, inplace=True)
raw_df.agg([missing_ratio])['Race']
```

```python
raw_df.fillna({'Sex': 'U'}, inplace=True)
raw_df.agg([missing_ratio])['Sex']
raw_df[13130:13141]['City']
```
